```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))

sns.countplot(x="Contract", hue="Churn", data=df, ax=axes[0,0])
axes[0,0].set_title("Contract vs Churn")

sns.countplot(x="InternetService", hue="Churn", data=df, ax=axes[0,1])
axes[0,1].set_title("Internet Service vs Churn")

sns.countplot(x="Dependents", hue="Churn", data=df, ax=axes[1,0])
axes[1,0].set_title("Dependents vs Churn")

sns.countplot(x="PhoneService", hue="Churn", data=df, ax=axes[1,1])
axes[1,1].set_title("Phone Service vs Churn")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True, common_norm=False, palette='coolwarm')
plt.title('Distribution of Monthly Charges against Churn Status')
plt.xlabel('Monthly Charges')
plt.ylabel('Density')
plt.show()
```

```python
num_columns = ['Contract',
               'InternetService',
               'Dependents',
               'PhoneService',
               'Churn']

df_encoded = df[num_columns].copy()

for col in df_encoded.columns:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes

corr_matrix = df_encoded.corr()
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            vmin=-1, vmax=1)
plt.title("Correlation Matrix")
plt.show()
```

### Critical Findings for customer contribute to "YES" amount of churn
1. Customer with "Electronics Check".
2. "Month-to-Month" contract customer.
3. Internet Service with "Fibre Optic"
4. With Phone Service.
