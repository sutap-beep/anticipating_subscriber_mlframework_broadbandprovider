```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/telco-customer-churn.csv")
```

```python
new_df = df.copy()
new_df["MonthlyCharges"] = new_df["MonthlyCharges"] * 1.2
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["MonthlyCharges"],
         bins=20,
         alpha=0.5,
         label="Original Data")

plt.hist(new_df["MonthlyCharges"],
         bins=20,
         alpha=0.5,
         label="Simulated New Data")

plt.title("Data Drift Analysis - Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```
