### Customer Churn Prediction - Business Understanding
1. Customer churn happens when customer stop using a company's service.
2. For a telecom company, losing customer reduces revenue and increases cost of acquiring new customer.
3. The goal of this project is to predict which customeres are likely to churn so the company can take preventive actions.

### Objective
To build a machine learning model that predicts customer churn.

### Business Impact
1. Improve customer retention.
2. Reduce revenue loss.
3. Create targeted marketing campaigns.
4. Improve customer satisfaction.

### Success Metrics
1. Accurately identify customeres likely to churn.
2. Provide useful business insights.
3. Help support customer retention strategies.
